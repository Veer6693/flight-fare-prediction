# Data Preprocessing

```python
train_df.head()
```

### Numrical features

#### Date_of_Journey

```python
train_df["Date_of_Journey"] = pd.to_datetime(train_df["Date_of_Journey"])
```

```python
train_df["Journey_Day"] = train_df["Date_of_Journey"].dt.day
```

```python
train_df["Journey_Month"] = train_df["Date_of_Journey"].dt.month
```

```python
train_df["Journey_DOW"] = train_df["Date_of_Journey"].dt.dayofweek
```

```python
train_df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x="Journey_Month", data=train_df)
```

```python
plt.figure(figsize=(6,3))
sns.countplot(x="Journey_DOW", data=train_df)
```

```python
max_prices = train_df.groupby("Journey_DOW")["Price"].mean()
plt.figure(figsize=(6, 3))
plt.scatter(max_prices.index, max_prices.values, color='b', marker='o')
plt.xlabel('Day of the Week')
plt.ylabel('Max Price')
plt.title('Max Price by Day of the Week')
plt.show()
```

#### Dep_Time

```python
train_df["Dep_Hour"] = pd.to_datetime(train_df["Dep_Time"]).dt.hour
```

```python
train_df["Dep_Minute"] = pd.to_datetime(train_df["Dep_Time"]).dt.minute
```

```python
train_df.drop("Dep_Time",axis=1,inplace=True)
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=train_df, x="Dep_Hour",bins=24)
```

#### Arrival_Time

```python
train_df["Arrival_Hour"] = pd.to_datetime(train_df["Arrival_Time"]).dt.hour
```

```python
train_df["Arrival_Minute"] = pd.to_datetime(train_df["Arrival_Time"]).dt.minute
```

```python
train_df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=train_df, x="Arrival_Hour",bins=24)
```

#### Duration

```python
duration = list(train_df["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) == 1:
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"
        else:
            duration[i] = "0h " + duration[i].strip()
```

```python
duration
```

```python
duration_hour = []
duration_minute = []
for i in duration:
    duration_hour.append(int(i.split(sep="h")[0]))
    duration_minute.append(int(i.split()[1].split(sep="m")[0]))
```

```python
train_df["Duration_Hour"] = duration_hour
```

```python
train_df["Duration_Minute"] = duration_minute
```

```python
train_df["Total_Duration(min)"] = (train_df["Duration_Hour"]*60)+(train_df["Duration_Minute"])
```

```python
train_df.drop("Duration",axis=1,inplace=True)
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=train_df, x="Total_Duration(min)",bins=30)
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=train_df, x="Total_Duration(min)", y="Price")
```

## Model Creation

```python
train_df.head()
```

```python
X = train_df.drop("Price",axis=1)
```

```python
y = train_df["Price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Lasso Regression": Lasso(alpha=1.0),
    "ElasticNet Regression": ElasticNet(alpha=1.0, l1_ratio=0.5),
    "SVR": SVR(kernel='rbf'),
    "Decision Tree Regression": DecisionTreeRegressor(),
    "Random Forest Regression": RandomForestRegressor(n_estimators=100),
    "Gradient Boosting Regression": GradientBoostingRegressor(n_estimators=100, learning_rate=0.1),
    "XGBoost Regression": XGBRegressor(n_estimators=100, learning_rate=0.1)
}

# Train and evaluate models
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test) 
    print(name)
    print(f"MAE: {mean_absolute_error(y_test, y_pred):.4f}")
    print(f"MSE: {mean_squared_error(y_test, y_pred):.4f}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.4f}")
    print(f"R^2 Score: {r2_score(y_test,y_pred):.4f}")
    print("-"*50)
```

## Hyperparameter Tuning 

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor
```

```python
xgbreg = XGBRegressor()
```

```python
param_grid = {
    'n_estimators': [550],
    'learning_rate': [0.01],
    'max_depth': [10],
    'min_child_weight': [2],
    'colsample_bytree': [0.9],
    'gamma': [0.01]
}
```

```python
grid_CV = GridSearchCV(estimator=xgbreg, param_grid=param_grid,cv=5,n_jobs=-1,verbose=1)
```

```python
grid_CV.fit(X_train,y_train)
```

```python
grid_CV.best_params_
```

```python
y_pred = grid_CV.predict(X_test)
```

```python
print(f"MAE: {mean_absolute_error(y_test, y_pred):.4f}")
print(f"MSE: {mean_squared_error(y_test, y_pred):.4f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.4f}")
print(f"R^2 Score: {r2_score(y_test,y_pred):.4f}")
```
